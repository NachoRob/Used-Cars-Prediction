## Análisis de Precios de Autos Usados mediante Modelos de Machine Learning

Un desafío clave en este sector es estimar de manera precisa el precio de un automóvil en función de sus características. Contar con un modelo predictivo eficiente puede beneficiar tanto a compradores como a vendedores, proporcionando una referencia objetiva de precios y ayudando a la toma de decisiones.

Este análisis tiene como objetivo principal desarrollar un modelo de Machine Learning que permita predecir con precisión el precio de un auto usado. Para lograr esto, exploraremos distintas variables disponibles en el dataset, evaluaremos su impacto en la determinación del precio y seleccionaremos el mejor modelo basado en métricas de rendimiento.

Para estructurar este análisis de manera rigurosa, aplicaremos la metodología **CRISP-DM (Cross-Industry Standard Process for Data Mining)**. Este enfoque nos permitirá organizar el trabajo en distintas fases, asegurando un proceso claro y sistemático.

#### Comparación de Modelos de Machine Learning para Predicción de Precios de Autos

| Modelo            | Descripción | Ventajas | Desventajas |
|------------------|------------|----------|-------------|
| **Regresión Lineal** | Modelo simple que asume una relación lineal entre las variables y el precio. | Fácil de interpretar, rápido de entrenar, útil como línea base. | No captura relaciones no lineales, puede ser sensible a outliers. |

#### **Manejo de Datos Atípicos (Outliers)**  
- Detectamos valores atípicos en las variables numéricas mediante técnicas como:  
  - **IQR (Rango Intercuartílico)** para eliminar valores extremos.  
  - **Isolation Forest** para detectar y filtrar outliers de forma automática.  

---

```python
# Creamos un dataset con las columnas a evaluar
numeric_cols = ["price_usd", "odometer_value", "year_produced", "engine_capacity"]

# Función para eliminar outliers con IQR ajustado
def remove_outliers_dynamic_iqr(df, columns):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1

        # Ajustar el factor de corte según la asimetría de la distribución
        skewness = df_clean[col].skew()
        factor = 1.5 if abs(skewness) < 1 else 2 if abs(skewness) < 2 else 2.5
        
        lower_bound = Q1 - factor * IQR
        upper_bound = Q3 + factor * IQR

        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    
    return df_clean

# Aplicar IQR ajustado
df_iqr_adjusted = remove_outliers_dynamic_iqr(df, numeric_cols)

# Comparar la cantidad de datos eliminados
print(f"Registros originales: {len(df)}")
print(f"Registros después de IQR ajustado: {len(df_iqr_adjusted)}")
print(f"Porcentaje de datos eliminados con IQR ajustado: {100 * (1 - len(df_iqr_adjusted) / len(df)):.2f}%")
```

```python
from sklearn.ensemble import IsolationForest
# Aplicar Isolation Forest

# Creamos dataset con las columnas a modificar
df_filtered = df[["price_usd", "odometer_value", "year_produced", "engine_capacity"]]

iso_forest = IsolationForest(contamination=0.08, random_state=42)  # Ajusta la contaminación según lo necesario
outliers = iso_forest.fit_predict(df_filtered)

# Filtrar los datos que no son outliers
df_iso = df_filtered[outliers == 1]

# Comparar la cantidad de datos eliminados
print(f"Registros originales: {len(df)}")
print(f"Registros después de Isolation Forest: {len(df_iso)}")
print(f"Porcentaje de datos eliminados con Isolation Forest: {100 * (1 - len(df_iso) / len(df)):.2f}%")
```

```python
# Comparar distribuciones antes y después de la eliminación de outliers
def compare_distributions(original_df, filtered_df, columns):
    fig, axes = plt.subplots(len(columns), 2, figsize=(12, len(columns) * 4))

    for i, col in enumerate(columns):
        # Antes de eliminar outliers
        sns.histplot(original_df[col], bins=50, kde=True, ax=axes[i, 0])
        axes[i, 0].set_title(f"Distribución de {col} (Antes de Isolation Forest)")

        # Después de eliminar outliers
        sns.histplot(filtered_df[col], bins=50, kde=True, ax=axes[i, 1])
        axes[i, 1].set_title(f"Distribución de {col} (Después de Isolation Forest)")

    plt.tight_layout()
    plt.show()

# Llamar a la función para comparar
compare_distributions(df, df_iso, numeric_cols)
```

###  **Escalado de Variables en el Dataset de Autos Usados**

En esta etapa del preprocesamiento, aplicamos **escalado de características** a las variables numéricas del dataset para mejorar el rendimiento de los modelos de Machine Learning. 

---

### **¿Por qué escalamos las variables?**
El escalado es importante porque:
- **Algunas variables tienen rangos muy diferentes** (Ej: `odometer_value` puede estar en cientos de miles, mientras que `engine_capacity` está en unidades pequeñas).
- **Los algoritmos de Machine Learning pueden ser sensibles a la escala de los datos**. Modelos como **Regresión Lineal** y **Gradient Boosting** se benefician de datos escalados.
- **Previene que una variable domine sobre las demás**, mejorando la estabilidad del modelo.

---

### **Métodos de escalado aplicados**
Seleccionamos distintos métodos según la distribución de cada variable:

| **Variable**         | **Método de escalado** | **Motivo** |
|----------------------|----------------------|------------|
| `odometer_value`    | **`RobustScaler`** | Resistente a outliers, ya que el kilometraje tiene valores extremos. |
| `year_produced`     | **`MinMaxScaler`**  | Normaliza los valores entre 0 y 1 sin distorsionar la escala temporal. |
| `engine_capacity`   | **`StandardScaler`** | Centra los datos en la media con desviación estándar 1, útil para distribuciones multimodales. |

---

```python
from sklearn.preprocessing import RobustScaler, MinMaxScaler, StandardScaler

# Creamos los escaladores según gráficos analizados en etapa EDA
scaler_odometer = RobustScaler()
scaler_year = MinMaxScaler()
scaler_engine = StandardScaler()


# Aplicamos los escaladores cada variable
df["odometer_value_scaled"] = scaler_odometer.fit_transform(df[["odometer_value"]])
df["year_produced_scaled"] = scaler_year.fit_transform(df[["year_produced"]])
df["engine_capacity_scaled"] = scaler_engine.fit_transform(df[["engine_capacity"]])

# Mostrar los primeros valores escalados para verificar
print(df[["odometer_value_scaled", "year_produced_scaled", "engine_capacity_scaled"]].head())
```

```python
# Crear copias para comparar antes y después del escalado
df_original = df.copy()

# Configurar los gráficos
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
fig.suptitle("Comparación de Distribuciones Antes y Después del Escalado", fontsize=16)

# Función para graficar antes y después
def plot_distribution(ax, data, title):
    sns.histplot(data, bins=50, kde=True, ax=ax)
    ax.set_title(title)

# Graficar las variables originales (izquierda) y escaladas (derecha)
plot_distribution(axes[0, 0], df_original["odometer_value"], "Distribución Original - odometer_value")
plot_distribution(axes[0, 1], df["odometer_value_scaled"], "Distribución Escalada - odometer_value")

plot_distribution(axes[1, 0], df_original["year_produced"], "Distribución Original - year_produced")
plot_distribution(axes[1, 1], df["year_produced_scaled"], "Distribución Escalada - year_produced")

plot_distribution(axes[2, 0], df_original["engine_capacity"], "Distribución Original - engine_capacity")
plot_distribution(axes[2, 1], df["engine_capacity_scaled"], "Distribución Escalada - engine_capacity")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

### Ajuste de Hiperparámetros con GridSearchCV  

En esta fase, optimizamos los modelos de **Machine Learning** ajustando sus hiperparámetros mediante **GridSearchCV**. Esto nos permite mejorar el desempeño de los modelos encontrando la mejor combinación de parámetros que vienen definidos "por defecto" en valores que no siempre pueden ser los óptimos. Esperamos mejorar el resultado de nuestras métricas y por ende de nuestros modelos. 

#### **Definición de Hiperparámetros a Evaluar**  
- **Random Forest:** Se ajustan la cantidad de árboles (`n_estimators`), la profundidad máxima (`max_depth`) y el número mínimo de muestras por nodo (`min_samples_split`).  
- **Gradient Boosting:** Se optimizan el número de estimadores (`n_estimators`), la tasa de aprendizaje (`learning_rate`) y la profundidad máxima (`max_depth`).  
- **Ridge Regression:** Se ajusta el parámetro de regularización (`alpha`).  

#### **Inicialización de los Modelos Base**  
- Creamos instancias de **RandomForestRegressor**, **GradientBoostingRegressor** y **Ridge** sin hiperparámetros predefinidos.  

#### **Configuración de GridSearchCV**  
- Se define **validación cruzada (`cv=3`)**, que divide los datos en tres subconjuntos para evaluar distintos hiperparámetros.  
- Se utiliza **`neg_mean_absolute_error`** como métrica de evaluación, minimizando el error absoluto medio.  
- Se ejecuta el ajuste en paralelo (`n_jobs=-1`) para mayor eficiencia.  

#### **Entrenamiento de Modelos con GridSearchCV**  
- Se entrenan los modelos con el **conjunto de entrenamiento (`X_train`, `y_train`)** evaluando todas las combinaciones de hiperparámetros.  

#### **Selección de los Mejores Hiperparámetros**  
- Se extraen los valores óptimos para cada modelo (`best_params_`) y se guardan para reutilizarlos en la fase final.  

#### **Entrenamiento de Modelos con los Mejores Hiperparámetros**  
- Se reentrenan los modelos utilizando los hiperparámetros óptimos encontrados en la búsqueda.  

#### **Evaluación de los Modelos Optimizados**  
- Se realizan predicciones sobre el conjunto de prueba (`X_test`).  
- Se calculan las métricas de desempeño:  
  - **MAE (Mean Absolute Error):** Error promedio en dólares.  
  - **MSE (Mean Squared Error):** Penaliza errores grandes.  
  - **R² Score:** Indica qué tan bien el modelo explica la variabilidad del precio.  

#### **Resultados del Ajuste de Hiperparámetros**  
- Se comparan los resultados de los modelos optimizados.  
- Se identifican mejoras respecto a la versión base, evaluando si el ajuste de hiperparámetros redujo el error de predicción.  

Este proceso permite seleccionar los mejores modelos y optimizar su desempeño en la predicción de precios de autos usados. 🚀  

### Ajustes de Hiperparámetros


```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor

# Definir hiperparámetros con opciones limitadas para evitar un GridSearch excesivo
param_grid_rf = {
    'n_estimators': [100, 300],  # Número de árboles
    'max_depth': [10, None],  # Profundidad del árbol
    'min_samples_split': [2, 5]  # División mínima en nodos
}

param_grid_gb = {
    'n_estimators': [100, 300],  
    'learning_rate': [0.05, 0.1],  
    'max_depth': [3, 5]  
}

param_grid_ridge = {
    'alpha': [0.1, 1, 10]  # Regularización en Ridge
}

param_grid_xgb = {
    'n_estimators': [100, 300],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5]
}

# Inicializar los modelos
rf = RandomForestRegressor(random_state=42)
gb = GradientBoostingRegressor(random_state=42)
ridge = Ridge()
xgb = XGBRegressor(objective='reg:squarederror', random_state=42)

# Configurar GridSearchCV
grid_search_rf = GridSearchCV(rf, param_grid_rf, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1, verbose=1)
grid_search_gb = GridSearchCV(gb, param_grid_gb, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1, verbose=1)
grid_search_xgb = GridSearchCV(xgb, param_grid_xgb, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1, verbose=1)

# Entrenar modelos con GridSearchCV
grid_search_rf.fit(X_train, y_train)
grid_search_gb.fit(X_train, y_train)
grid_search_xgb.fit(X_train, y_train)

# Obtener mejores hiperparámetros
best_params_rf = grid_search_rf.best_params_
best_params_gb = grid_search_gb.best_params_
best_params_xgb = grid_search_xgb.best_params_

# Entrenar modelos con los mejores hiperparámetros
best_rf = RandomForestRegressor(**best_params_rf, random_state=42)
best_gb = GradientBoostingRegressor(**best_params_gb, random_state=42)
best_xgb = XGBRegressor(**best_params_xgb, objective='reg:squarederror', random_state=42)

models = {
    "Random Forest (Optimizado)": best_rf,
    "Gradient Boosting (Optimizado)": best_gb,
    "XGBoost (Optimizado)": best_xgb
}

# Evaluar modelos optimizados
results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    results.append([name, mae, mse, r2])

# Crear DataFrame con los resultados optimizados
df_results = pd.DataFrame(results, columns=["Modelo", "MAE", "MSE", "R² Score"])

# Mostrar los mejores hiperparámetros y resultados optimizados
print("Mejores Hiperparámetros - Random Forest:")
print(pd.DataFrame([best_params_rf]))
print("\nMejores Hiperparámetros - Gradient Boosting:")
print(pd.DataFrame([best_params_gb]))
print("\nMejores Hiperparámetros - XGBoost:")
print(pd.DataFrame([best_params_xgb]))
print("\nResultados del Modelo Optimizado:")
print(df_results)
```
